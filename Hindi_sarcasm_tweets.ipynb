```python
def processText(text):
    text = text.lower()
    text = re.sub(r'((www\.[^\s]+)|(https?://[^\s]+))', '', text)  # Remove URLs
    text = re.sub(r'@[^\s]+', '', text)  # Remove @mentions
    text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces with a single space
    text = re.sub(r'#([^\s]+)', r'\1', text)  # Remove the # symbol from hashtags
    text = re.sub(r'[.!:?\-\'"/]', '', text)  # Remove punctuation
    text = text.strip('"')
    return text

# Apply processText function to each row in the 'text' column
for i in range(len(df)):
    df.loc[i, 'text'] = processText(df.loc[i, 'text'])

# Function to count the length of the text
def count_length():
    df['word_count'] = df['text'].apply(lambda x: len(str(x).split(" ")))


count_length()
df.tail(10)
```

```python
from sklearn.model_selection import KFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Handle NaN or empty values in the text column
df['text'].fillna('', inplace=True)

# Define features (X) and labels (y)
X = df['text'].astype(str)  # Ensure X is a Series of strings
y = df['label']

# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf = vectorizer.fit_transform(X)

# Initialize K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize model
nb_model = MultinomialNB()

# Lists to store metrics for each fold
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []

# Perform K-Fold Cross-Validation
for train_index, test_index in kf.split(X_tfidf):
    X_train, X_test = X_tfidf[train_index], X_tfidf[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Ensure that y_train and y_test are in correct format (integer encoded if necessary)
    y_train = np.array(y_train)
    y_test = np.array(y_test)

    # Train the model
    nb_model.fit(X_train, y_train)

    # Make predictions
    y_pred = nb_model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Append metrics to lists
    accuracy_list.append(accuracy)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)

# Calculate and print average metrics
print(f'Average Accuracy: {np.mean(accuracy_list) * 100:.2f}%')
print(f'Average Precision: {np.mean(precision_list) * 100:.2f}%')
print(f'Average Recall: {np.mean(recall_list) * 100:.2f}%')
print(f'Average F1-score: {np.mean(f1_list) * 100:.2f}%')

```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Handle NaN or empty values in the text column
df['text'].fillna('', inplace=True)

# Define features (X) and labels (y)
X = df['text'].astype(str)  # Ensure X is a Series of strings
y = df['label']

# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf = vectorizer.fit_transform(X)

# Initialize Stratified K-Fold Cross-Validation
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Lists to store metrics for each fold
accuracies = []
precisions = []
recalls = []
f1s = []

# Perform Stratified K-Fold Cross-Validation
for train_index, test_index in kfold.split(X_tfidf, y):
    X_train, X_test = X_tfidf[train_index], X_tfidf[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Train the Naive Bayes model
    nb_model = MultinomialNB()
    nb_model.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = nb_model.predict(X_test)

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    # Store metrics
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1s.append(f1)

# Compute and print average metrics
print(f'Average Accuracy: {np.mean(accuracies) * 100:.2f}%')
print(f'Average Precision: {np.mean(precisions) * 100:.2f}%')
print(f'Average Recall: {np.mean(recalls) * 100:.2f}%')
print(f'Average F1-score: {np.mean(f1s) * 100:.2f}%')
```

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ensure all text is in string format and fill NaN values with empty strings
df['text'].fillna('', inplace=True)

# Vectorize the text data using TF-IDF (this turns the text into numerical vectors)
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf = vectorizer.fit_transform(df['text'].astype(str))

# Apply KMeans clustering
k = 2  # Since there are two classes: sarcastic and non-sarcastic
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X_tfidf)

# Add the predicted cluster labels to the dataframe
df['cluster'] = kmeans.labels_

# Visualize the clusters (using the first 2 principal components for visualization)
from sklearn.decomposition import PCA

# Reduce dimensions to 2 for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_tfidf.toarray())

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['cluster'], cmap='viridis', s=50)
plt.title('K-Means Clustering of Sarcasm vs Non-Sarcasm Tweets')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()

# Display the first few rows of the dataframe with clusters
print(df[['text', 'cluster']].head())

# Optionally, check the centroid of each cluster
print("Cluster Centroids:")
print(kmeans.cluster_centers_)

# Get the most frequent words in each cluster
def get_top_keywords(tfidf_matrix, top_n=10):
    # Get the feature names (terms) corresponding to the columns of the tfidf matrix
    feature_names = np.array(vectorizer.get_feature_names_out())

    # Get the top n words per cluster
    top_keywords = {}
    for i, centroid in enumerate(kmeans.cluster_centers_):
        top_idx = centroid.argsort()[-top_n:][::-1]  # Get the top n words
        top_keywords[i] = feature_names[top_idx]

    return top_keywords

# Display top words for each cluster
top_keywords = get_top_keywords(X_tfidf, top_n=10)
print("\nTop keywords in each cluster:")
for cluster, keywords in top_keywords.items():
    print(f"Cluster {cluster}: {', '.join(keywords)}")
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns

# Assuming df['text'] contains the preprocessed text and we have already applied TF-IDF Vectorization.
# If not done, here is the vectorization again:
from sklearn.feature_extraction.text import TfidfVectorizer

# Vectorize the text data using TF-IDF
df['text'] = df['text'].apply(lambda x: ' '.join(x))  # Join tokenized words back to a string
vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 2))
X_tfidf = vectorizer.fit_transform(df['text'])

# --- 1. Agglomerative Hierarchical Clustering ---
# Using Cosine Distance as the metric for AgglomerativeClustering
agg_clust = AgglomerativeClustering(n_clusters=2, affinity='cosine', linkage='average')
df['agg_labels'] = agg_clust.fit_predict(X_tfidf.toarray())

# Display Agglomerative Clustering Results
print("Agglomerative Clustering Results:")
print(df[['text', 'agg_labels']].head())

# --- 2. Gaussian Mixture Model (GMM) ---
# GMM assumes the data is drawn from a mixture of several Gaussian distributions
gmm = GaussianMixture(n_components=2, random_state=42)
df['gmm_labels'] = gmm.fit_predict(X_tfidf.toarray())

# Display Gaussian Mixture Model Results
print("GMM Clustering Results:")
print(df[['text', 'gmm_labels']].head())

# --- 3. Visualization ---
# Reduce the dimensionality using PCA for 2D visualization
pca = PCA(n_components=2)
pca_components = pca.fit_transform(X_tfidf.toarray())

# Visualize Agglomerative Clustering Results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=df['agg_labels'], palette="Set2", s=100, edgecolor='black')
plt.title("Agglomerative Hierarchical Clustering Results")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()

# Visualize GMM Clustering Results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=df['gmm_labels'], palette="Set2", s=100, edgecolor='black')
plt.title("Gaussian Mixture Model Clustering Results")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```
